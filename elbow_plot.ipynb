```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pickle
from kneed import KneeLocator
```

```python
path = 'D:\GitHub\Clone\planning\saved_centers'
```

```python
path_list = os.listdir(path)
```

```python
elbow_file_3_to_5 = path_list[30:60]
print(elbow_file_3_to_5)
```

```python
elbow = []
best_in_a_k = []

for i in range(3):
    for file in elbow_file_3_to_5[i*10:(i+1)*10]:
        km_dba = pickle.load(open("D:\\GitHub\\Clone\\planning\\saved_centers\\"+file,'rb'))
        best_in_a_k.append(km_dba.inertia_)
    elbow.append(min(best_in_a_k))   
```

```python
elbow_6_12 = open("D:\\GitHub\\Clone\\planning\\elbow_list.pkl",'rb')
```

```python
elbow_value_6_to_12 = pickle.load(elbow_6_12)
print(elbow_value_6_to_12)
```

```python
k_min = 3
k_max = 12
plt.figure(figsize=(12,10))
plt.plot(range(k_min, k_max+1), elbow+elbow_value_6_to_12, 'c*-')
plt.xlabel('Number of centroids',fontsize=20)
plt.ylabel('Elbow score',fontsize=20)
plt.title("Elbow score on vehicle features",fontsize=20)
plt.xticks(range(3,13),fontsize=15)
kn = KneeLocator(range(3,13), elbow+elbow_value_6_to_12, curve='convex', direction='decreasing')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed',color='m')
plt.savefig("D:\\GitHub\\Clone\\planning\\Full Elbow plot")
plt.show
```

```python
pickle.dump(elbow+elbow_value_6_to_12, open("D:\\GitHub\\Clone\\planning\\full_elbow_list.pkl", 'wb'))
```
